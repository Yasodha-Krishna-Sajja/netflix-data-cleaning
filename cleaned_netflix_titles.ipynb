# Netflix Data Cleaning Project

## Objective
To clean and preprocess raw Netflix dataset by handling missing values, duplicates, and inconsistent formats.

```python
import pandas as pd

# load dataset
df = pd.read_csv("netflix_titles.csv")
```

```python
df.info()
df.isnull().sum()
```

The dataset contains missing values in director, cast, country, date_added, rating, and duration columns.

```python
# Handling missing values
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Not Specified')
```

```python
# Dropping small missing values
df = df.dropna(subset=['date_added'])
```

```python
# Filling small missing values
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Clean column names
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
# remove extra spaces
df['date_added'] = df['date_added'].str.strip()
```

```python
# Convert date format
df['date_added'] = pd.to_datetime(df['date_added'])
```
